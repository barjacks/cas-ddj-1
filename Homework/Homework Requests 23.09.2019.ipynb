```python
# Version 3

url = 'https://earthquake.usgs.gov/fdsnws/event/1/count?starttime=2014-09-23&endtime=2019-09-23&minmagnitude=5.5'
response = requests.get(url)
response.text
```

```python
#Version 2
#<- nur des letzten Monats, weil es sonst mein Jupyter Notebook sprengt.
url = 'https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=2019-08-22&endtime=2019-09-23&minmagnitude=5.5'
r = requests.get(url)
r = r.json()

erdbeben = r['features']

eine_liste = []
for item in erdbeben:
    magnitude = item['properties']['mag']
    place = item['properties']['place']
    
    mini_dict = {"Mangitude": magnitude, 
                "Place": place}
    eine_liste.append(mini_dict)
pd.DataFrame(eine_liste).head()
```

```python
# Version 1

url = 'https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=2019-08-22&endtime=2019-09-23'
r = requests.get(url)
r = r.json()

erdbeben = r['features']
```

```python
# real stuff <- nur des letzten Monats, weil es sonst mein Jupyter Notebook sprengt.

mag_lst = []

for item in erdbeben:
    #print(item['properties']['mag'])
    try:
        if float(item['properties']['mag']) >= 5.5:
            magnitude = item['properties']['mag']
            place = item['properties']['place']
        #zeit = item[['properties']['time']] <- fÃ¼r spÃ¤ter

            mini_dict = {"Place": place,
                    "Magnitude": magnitude}
            mag_lst.append(mini_dict)
    except:
        print("The following event could not be processed: \n", item['properties'])

pd.DataFrame(mag_lst).head()
```

4. Zeige eine Liste der 10 stÃ¤rksten Erdbeben des laufenden Jahres an. (Nicht vergessen, nun musst Du die Methode wechseln. Nicht mehr "count", sondern "query").

```python
# real stuff 

max_lst = []

for item in erdbeben:
    magnitude = item['properties']['mag']
    place = item['properties']['place']
    
    mini_dict = {"Mangitude": magnitude, 
                "Place": place}
    max_lst.append(mini_dict)

pd.DataFrame(max_lst)
```

6. Ziehe auch die Zeit mit heraus?

```python
max_lst = []

for item in erdbeben:
    magnitude = item['properties']['mag']
    place = item['properties']['place']
    zeit =      (item['properties']['time'])
    zeit = zeit/1000
    zeit =  datetime.utcfromtimestamp(zeit).strftime('%Y-%m-%d') #time.ctime(int(zeit))

    
    mini_dict = {"Mangitude": magnitude, 
                "Place": place,
                 "Time": zeit}
    
    max_lst.append(mini_dict)

pd.DataFrame(max_lst)
```
